```python
import bolt
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = bolt.Cosmo(h=0.7, Omega_m=0.3, Omega_b=0.049, A_s=2.1e-9, n_s=0.96)
bolt.calc_background(c)
z = np.linspace(0, 100, 100)
D_L = bolt.get_luminosity_distances(z)
plt.loglog(z, D_L)
plt.xlabel("z")
plt.ylabel("$D_L(z)$ [Mpc]")
```
